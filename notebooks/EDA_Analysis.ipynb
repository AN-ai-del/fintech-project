# Key EDA Findings

1. NAV values showed long-term growth across most schemes.
2. SBI Mutual Fund maintained the highest AUM.
3. SIP inflows reached record levels by the end of 2025.
4. Equity categories attracted higher inflows than debt categories.
5. Investor participation was strongest in the 26–35 age group.
6. Male investors accounted for the majority of transactions.
7. T30 cities generated most mutual fund transactions.
8. Folio counts nearly doubled between 2022 and 2025.
9. Fund returns showed moderate positive correlation.
10. Banking, Financial Services and Technology sectors dominated portfolio allocations.
