# Advanced Analytics + Risk Metrics

This notebook summarizes Day 6 advanced analytics for the Mutual Fund Analytics Capstone Project.

## 1. Historical VaR and CVaR

Historical VaR at 95% confidence was calculated using the 5th percentile of daily returns. CVaR was calculated as the average of returns below the VaR threshold.

```
import pandas as pd
var_cvar = pd.read_csv('../data/processed/var_cvar_report.csv')
var_cvar.sort_values('var_95').head(10)
```

## 2. Rolling 90-Day Sharpe Ratio

Rolling Sharpe Ratio tracks the changing risk-adjusted performance of selected funds over time.

```
from IPython.display import Image
Image(filename='../reports/charts/rolling_sharpe_chart.png')
```

## 3. Investor Cohort Analysis

Investors were grouped by the year of their first transaction to analyze cohort-level behavior.

```
cohort = pd.read_csv('../data/processed/investor_cohort_analysis.csv')
cohort
```

## 4. SIP Continuity Analysis

Investors with 6 or more SIP transactions were analyzed for average gap between SIP dates. Investors with gaps greater than 35 days were flagged as at-risk.

```
sip_continuity = pd.read_csv('../data/processed/sip_continuity_report.csv')
sip_continuity['risk_flag'].value_counts()
```

## 5. Fund Recommender

A simple rule-based recommender was created using risk appetite and Sharpe Ratio ranking.

```
recommendations = pd.read_csv('../data/processed/fund_recommendations.csv')
recommendations
```

## 6. Sector HHI Concentration

Sector concentration was measured using HHI = sum of squared sector weights.

```
hhi = pd.read_csv('../data/processed/sector_hhi_report.csv')
hhi.head(10)
```

## 7. Advanced Insights

1. Funds with the most negative VaR values have the highest downside risk exposure.
2. CVaR gives a stricter view of tail losses compared to VaR.
3. Rolling Sharpe Ratio helps identify periods of unstable risk-adjusted performance.
4. SIP continuity analysis identifies investors who may be at risk of discontinuing systematic investments.
5. Sector HHI highlights funds with concentrated portfolio exposure.
